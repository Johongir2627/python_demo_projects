## Multiprocessing

```python
import multiprocessing
import time

def task(name):
    print(f"Starting process {name}")
    time.sleep(2)
    print(f"Finished process {name}")

start = time.time()

processes = []
for i in range(3):
    p = multiprocessing.Process(target=task, args=(i,))
    processes.append(p)
    p.start()

for p in processes:
    p.join()

end = time.time()
print(f"Total time (multiprocessing): {end - start:.2f} seconds")
```

```python
def flatten_list(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten_list(item))  # recursive call
        else:
            result.append(item)
    return result

# Example
a = [1, [2, [3, 4], 5], 6]
flattened = flatten_list(a)
print(flattened)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
a = [1, [2, [3, 4], 5], 6]
b = []
for i in a:
    if type(i) == int:
        b.append(i)
    else:
        for j in i:
            if type(j) == int:
               b.append(j)
            else:
                for k in j:
                    if type(k) == int:
                        b.append(k)


print(b)
```
