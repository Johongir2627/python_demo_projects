```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

plt.scatter(X_data, y_data)
plt.show()
```

```python
plt.scatter(X_data, y_data, c = 'red', marker='*')
plt.show()
```

```python
plt.scatter(X_data, y_data, c = 'red', marker='*', s = 160)
plt.show()
```

```python
X_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100

plt.scatter(X_data, y_data, c = 'red', marker='*', s = 160, alpha=0.3)
plt.show()
```

```python
years = [2006 + x for x in range(8)]
weights = [81, 80, 91, 76, 82, 88, 79, 79]

plt.plot(years, weights, c = 'g', lw = 3, linestyle = '--')
```

```python
plt.plot(years, weights, 'b--', lw = 3)
```

```python
x = ['Ferrari', 'Bugatti', 'Fiat', 'Volvo']
y = [200, 320, 180, 160]

plt.bar(x, y, color='red')
plt.show()
```

```python
plt.bar(x, y, color='red', align='edge')
plt.show()
```

```python
plt.bar(x, y, color='red', align='edge', width=0.5)
plt.show()
```

```python
plt.bar(x, y, color='red', align='edge', width=0.5, edgecolor = 'black', lw = 6)
plt.show()
```

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages)
plt.show()
```

```python
plt.hist(ages, bins = 15)
plt.show()
```

```python
plt.hist(ages, bins = [ages.min(), 18, 21, ages.max()])
plt.show()
```

```python
ages = np.random.normal(20, 7.5, 1000)
plt.hist(ages, bins = 20)
plt.show()
```

```python
ages = np.random.normal(20, 7.5, 1000)
plt.hist(ages, bins = 20, cumulative=True)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'C#']
votes = [200, 50, 80, 30]

plt.pie(votes, labels=langs)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'C#']
votes = [200, 50, 80, 30]
explodes = [0.3, 0, 0, 0]

plt.pie(votes, labels=langs, explode=explodes)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'C#']
votes = [200, 50, 80, 30]
explodes = [0.1, 0.1, 0.1, 0.1]

plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%')
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'C#']
votes = [200, 50, 80, 30]
explodes = [0.1, 0.1, 0.1, 0.1]

plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=0.8)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'C#']
votes = [200, 50, 80, 30]
explodes = [0.1, 0.1, 0.1, 0.1]

plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=1.3)
plt.show()
```

```python
langs = ['Python', 'Java', 'C++', 'C#']
votes = [200, 50, 80, 30]
explodes = [0.1, 0.1, 0.1, 0.1]

plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=1.3, startangle=90)
plt.show()
```
