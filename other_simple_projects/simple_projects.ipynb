```python
# 3.7 Print reverse number pattern
n = int(input("Enter a number: "))
i = 0
while i < n:
    j = 0
    while j < n - i:
        print(n - j - i, end=' ')
        j += 1
    print()
    i += 1
```

```python
# 3.11 Print all prime numbers within a range
for i in range(5, 25):
    for j in range(2, i):
        if i%j==0:
            break
    else:
        print(i)
```

```python
# 3.12 Display Fibonacci series up to 10 terms
n = 10
a = 0
b = 1
print(a, b, end=' ')
for i in range(2, n):
    c = a + b
    print(c, end=' ')
    a = b
    b = c
    
```
