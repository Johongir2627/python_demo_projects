```python
# How to find average of n numbers in puthon

numbers = []
while True:
    n = input("Enter numbers to find average, if you want to stop enter 'q': ")
    try:
        n = float(n)
        numbers.append(n)
    except:
        if n == 'q':
            print("Exiting the program.")
            break
    else:
        print("Invalid input. Please enter a number or 'q' to quit.")

average = sum(numbers) / len(numbers)
print(f"The average of the numbers is: {average:.2}")
```

```python
# How to sum of the first n positive integers in python
n = int(input("Enter a positive integer: "))

sum = (n * (n + 1)) / 2
print(sum)
```
