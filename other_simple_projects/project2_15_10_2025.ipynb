```python
#1. Find customers who haven't placed any orders

import pandas as pd
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'customer_name': ['Alice', 'Bob', 'Charlie', 'David']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 1]
})

merged = customers.merge(orders, left_on = 'customer_id', right_on = 'customer_id', how = 'left')
merged[merged['order_id'].isna()]['customer_name']
```

```python
#2. Find sudden temperature drops, use diff() function
df = pd.DataFrame({
    'date': pd.date_range(start='2023-01-01', periods=7),
    'temperature': [30, 28, 25, 20, 22, 18, 15]
})

df['temp_diff'] = df['temperature'].diff()
sudden_drops = df[df['temp_diff'] <= -5]
print(sudden_drops[['date', 'temperature', 'temp_diff']])
```
