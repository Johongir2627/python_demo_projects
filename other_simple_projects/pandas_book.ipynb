```python
import pandas as pd
import numpy as np

# Create series from numpy array
v = np.array([1, 2, 3, 4, 5, 6, 7])
s1 = pd.Series(v)

# Get the memory usage of the series
memory_usage = s1.memory_usage(deep=True)
print(f"Memory usage of the series: {memory_usage} bytes")

# Get the number of bytes allocated to each item
itemsize = s1.memory_usage(deep=True) / len(s1)
print(f"Number of bytes allocated to each item: {itemsize} bytes")
```
