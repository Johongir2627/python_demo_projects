```python
import pandas as pd
customers = pd.DataFrame({
    'customer_id1': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 1]
})
```

```python
df = pd.merge(customers, orders, left_on='customer_id1', right_on='customer_id', how='left')
df[df['order_id'].isna()]['name']
df = df.drop(columns=['customer_id'])
df
```

```python
#Find sudden temperature drops, use diff() function
df = pd.DataFrame({
    'date': pd.date_range(start='2024-01-01', periods=7),
    'temperature': [30, 28, 27, 15, 14, 25, 10]
})

df['temp_diff'] = df['temperature'].diff()
df['sudden_drop'] = df['temp_diff'] < -10
df
```
