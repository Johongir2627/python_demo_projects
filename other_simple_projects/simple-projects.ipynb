```python
# i calculator
# This script performs basic arithmetic operations: addition, subtraction, multiplication, and division.

operator = input("Enter the operator (+, -, *, /): ")
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    result = num1 / num2
else:
    print("Invalid operator!")
    exit()

print(round(result, 2))
```

```python
# Python weight converter

weight = float(input("Enter your weights: "))
unit = input("Kilogramm or Pountds (kg/lb): ").lower()

if unit == "kg":
    weight = weight * 2.20462
    units = "lb"
elif unit == "lb":
    weight = weight / 2.20462
    units = "kg"
else:
    print("Invalid unit!")
    exit()

print(f'Your weight is {round(weight, 2)} {units}')
```

```python
# Temperature converter
temp = float(input("Enter the temperature: "))
unit = input("Celsius or Fahrenheit (C/F): ").lower()

if unit == "c":
    temp = (temp * 9/5) + 32
    units = "F"
elif unit == "f":
    temp = (temp - 32) * 5/9
    units = "C"
else:
    print(f'{unit} is not a valid unit')
    exit()

print(f'The temperature is {round(temp, 2)}°{units}')
```

```python
# Compound interest calculater
# A = P(1 + r/n)^(nt)
# A = the future value of the investment/loan, including interest
# P = the principal investment amount (the initial deposit or loan amount)
# r = the annual interest rate (decimal)
# n = the number of times that interest is compounded per unit t
# t = the time the money is invested or borrowed for, in years

principal = 0
rate = 0
time = 0


while principal <= 0:
    principal = float(input("Enter the principal amount (P): "))
    if principal <= 0:
        print("Principal amount must be greater than 0. Please try again.")


while True:
    rate = float(input("Enter the interest rate: "))
    if rate <= 0:
        print("Interest rate must be greater than 0. Please try again.")
    else:
        break

while time <= 0:
    time = float(input("Enter the time in years: "))
    if time <= 0:
        print("Time must be greater than 0. Please try again.")


total = principal * (1 + rate / 100) ** time
print(f"The total amount after {time} years is: ${total:.2f}")


print(f"Principal amount: {principal}")
print(f"Interest rate: {rate}")
print(f"Time in years: {time}")
```

```python
# Countdown timer
import time
my_time = int(input("Enter the time in seconds: "))

for x in range(my_time, 0, -1):
    seconds = x % 60
    minutes = (x // 60) % 60
    hours = (x // 3600)
    print(f"{hours:02}:{minutes:02}:{seconds:02}", end="\r")
    time.sleep(1)

print("Time\'s up")
```

```python
# Shopping Card Program
foods = []
prices = []
total = 0

while True:
    food  = input("Enter the food item (or 'q' to finish): ").lower()
    if food.lower() == 'q':
        break
    else:
        price = float(input(f"Enter the price of {food}: $"))
        foods.append(food)
        prices.append(price)






print("----- Shopping -----")
for food in foods:
    index = foods.index(food)
    price = prices[index]
    print(f"{food.capitalize()}: ${price:.2f}")
    total += price

# making horizontal
for food in foods:
    print(food, end=" ")


for price in prices:
    total += price

print(f"\nTotal: ${total:.2f}")
```
