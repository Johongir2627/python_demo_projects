```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Choose your custom background color here
custom_bg = "#23272f"  # Change this to any color you want

data = {
    "Yillar": [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Y": [45.10, 41.50, 38.90, 37.90, 37.30, 35.00, 34.80, 34.80],
    "X1": [60.9, 58.6, 57.7, 42.8, 39.6, 40.0, 39.0, np.nan],
    "X2": [26.56, 26.52, 25.48, 22.25, 19.73, 22.32, 21.50, 19.25],
    "X3": [0.065, 0.054, 0.040, 0.040, 0.058, 0.025, 0.048, 0.029],
    "X4": [22, 23, 25, 26, 28, 31, 33, 32],
    "X5": [13011, 8605, 7824, 7514, 12988, 17559, 17775, np.nan],
    "X6": [10.40, 12.70, 11.30, 13.40, 13.40, 12.00, 11.60, np.nan],
    "X7": [5.83, 9.30, 8.98, 10.50, 9.60, 6.95, 6.80, 4.50],
    "X8": [13.90, 17.50, 14.50, 12.90, 10.80, 11.40, 10.00, 9.60]
}
df = pd.DataFrame(data)

X_vars = ["X1", "X2", "X3", "X4", "X5", "X6", "X7", "X8"]

for x in X_vars:
    df_simple = df.dropna(subset=[x, "Y"])
    X = df_simple[[x]]
    Y = df_simple["Y"]
    X_const = sm.add_constant(X)
    model = sm.OLS(Y, X_const).fit()
    fig, ax = plt.subplots(figsize=(8,5))
    fig.patch.set_facecolor(custom_bg)      # Figure background
    ax.set_facecolor(custom_bg)             # Axes background
    ax.scatter(df_simple[x], Y, color="#f5f6fa", alpha=0.85, s=60, edgecolor=None)  # smaller dots
    ax.plot(df_simple[x], model.predict(X_const), color="#ffe600", linewidth=3)
    ax.set_xlabel(f"{x}", fontsize=20, weight='bold', color='white')  # bigger text
    ax.set_ylabel("Y", fontsize=20, weight='bold', color='white')     # bigger text
    ax.set_title(f"Linear Regression: Y vs {x}", fontsize=22, weight='bold', color='white')
    ax.grid(False)
    # Make x and y axes lines white and thicker
    ax.spines['bottom'].set_color('white')
    ax.spines['left'].set_color('white')
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.tick_params(axis='x', colors='white', labelsize=13)
    ax.tick_params(axis='y', colors='white', labelsize=13)
    plt.tight_layout()
```
