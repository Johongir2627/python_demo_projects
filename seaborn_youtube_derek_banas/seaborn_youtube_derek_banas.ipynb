```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

## Distribution Plots

```python
sns.histplot(crash_df['not_distracted'], kde=True)
```

```python
sns.histplot(crash_df['not_distracted'], kde=False)
```

```python
sns.histplot(crash_df['not_distracted'], kde=False, bins=20)
```

## Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

## KDE plot

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.kdeplot(crash_df['alcohol'], bw_adjust=0.5, fill=True, color='red', alpha=0.5)
```

```python
sns.pairplot(crash_df)
```
