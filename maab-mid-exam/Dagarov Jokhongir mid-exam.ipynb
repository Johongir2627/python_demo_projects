```python
#4. Print reverse number pattern
# 5 4 3 2 1
# 4 3 2 1
# 3 2 1
# 2 1
# 1

n = 5
i = 0
while i <= n:
    j = 0
    while j < n-i:
        print(n-i-j, end = ' ')
        j+=1
    print()
    i+=1
```

```python
#5. Write a function even_odd_sort(numbers) that separates even and odd numbers from a list.
#Return a tuple where the first element is a sorted list of even numbers, and the second is a sorted list of odd numbers.

def even_odd_sort(numbers):
    my_list = [i for i in range(1, numbers+1)]
    even = []
    odd = []
    for j in my_list:
        if j % 2 == 0:
            even.append(j)
        else:
            odd.append(j)
    last_list = [sorted(even), sorted(odd)]
    return tuple(last_list)


even_odd_sort(15)



```

```python
#8. Create a function for a Number Guessing Game where the computer randomly selects a number between 1 and 100, 
# and the player has to guess it.

# Requirements:
# Display a welcome message and inform the player about the game rules.+
# Randomly generate a number between 1 and 100.+
# Allow the player to choose a difficulty level:+
# "easy" → 10 attempts
# "hard" → 5 attempts
# Give hints after each guess:
# If the guess is too high, print "Too high."
# If the guess is too low, print "Too low."
# If the guess is correct, print "You got it! The answer was <number>." and end the game.
# End the game if the player runs out of attempts, displaying "You've run out of guesses, you lose."
# Handle invalid difficulty inputs by displaying "Invalid input" and exiting the game.
import random



def number_guessing_game():
    while True:
        print('''****************************\nGame rules: \nChoose number between 1 and 100. \nIf the number you choose is the same with computer choose you win. \nYou have 2 difficulty level: easy and hard\n****************************''')
        difficulty_level = input("Choose difficulty level: easy → 10 attempst / hard → 5 attempts")
        if difficulty_level == 'easy':
            a = 1
            computer_selected_number = random.randint(1, 100)
            invalid_attempts = []
            while a <= 10:
                a += 1
                player_guessed_number = int(input("Enter an integer number:"))
                if computer_selected_number > player_guessed_number:
                    print("Too low")
                    invalid_attempts.append(player_guessed_number)
                elif computer_selected_number < player_guessed_number:
                    print('Too high')
                    invalid_attempts.append(player_guessed_number)
                elif computer_selected_number == player_guessed_number:
                    print(f'The answer was {player_guessed_number}!')
                    break
                else:
                    invalid_attempts.append(str(player_guessed_number))
                    
            else:
                print(f'You lost. Answer was {computer_selected_number}')
                print(f'Invalid attemps: {invalid_attempts}')
                break
            break
        elif difficulty_level == 'hard':
            a = 1
            computer_selected_number = random.randint(1, 100)
            invalid_attempts = []
            while a <= 5:
                a += 1
                player_guessed_number = int(input("Enter an integer number:"))
                if computer_selected_number > player_guessed_number:
                    print("Too low")
                    invalid_attempts.append(player_guessed_number)
                elif computer_selected_number < player_guessed_number:
                    print('Too high')
                    invalid_attempts.append(player_guessed_number)
                elif computer_selected_number == player_guessed_number:
                    print(f'The answer was {player_guessed_number}!')
                    break
                else:
                    invalid_attempts.append(str(player_guessed_number))
                    
            else:
                print(f'You lost. Answer was {computer_selected_number}')
                print(f'Invalid attemps: {invalid_attempts}')
                break
            break
        else:
            print('Invalid input')
        

    
number_guessing_game()

```
