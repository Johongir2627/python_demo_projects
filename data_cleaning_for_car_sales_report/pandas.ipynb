```python
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv('car_prices.csv')
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
df = df[df['make'].notna() & (df['make'].str.strip() != '')] #drop null values in 'make' column
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
df = df[df['model'].notna() & (df['model'].str.strip() != '')]  # drop null or blank values in 'model' column
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
# Replace blank or null values in 'trim' and 'body' columns with "Other"
df['trim'] = df['trim'].fillna('').replace(r'^\s*$', 'Other', regex=True)
df['trim'] = df['trim'].replace('', 'Other')

df['body'] = df['body'].fillna('').replace(r'^\s*$', 'Other', regex=True)
df['body'] = df['body'].replace('', 'Other')
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
print(df[df['transmission'].str.lower() != 'automatic'])
```

```python
df['transmission'] = df['transmission'].replace(r'^\s*$', np.nan, regex=True)
```

```python
df['transmission'] = df['transmission'].replace(r'^\s*$', np.nan, regex=True)
```

```python
df['transmission'] = df['transmission'].apply(
    lambda x: x if str(x).strip().lower() in ['automatic', 'manual'] else np.nan
)
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
for col in ['vin', 'state', 'color', 'interior', 'seller']:
    df[col] = df[col].fillna('').replace(r'^\s*$', 'Other', regex=True)
    df[col] = df[col].replace('', 'Other')
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
for col in ['condition', 'odometer', 'mmr', 'sellingprice']:
    df[col] = pd.to_numeric(df[col], errors='coerce') 
    mean_val = df[col].mean()
    df[col] = df[col].fillna(mean_val)
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```

```python
df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce')
df = df[df['saledate'].notna()]
df['saledate'] = df['saledate'].dt.strftime('%d-%m-%Y %H:%M:%S')
```

```python
number_of_rows = len(df)
print("Number of rows:", number_of_rows)
```
