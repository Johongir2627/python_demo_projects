```python
import pandas as pd
import json


# convert to DataFrame
df = pd.DataFrame(json_data["data"])

# set float format like 2511.00
pd.set_option('display.float_format', '{:.2f}'.format)

# display DataFrame
print(df)
```

```python
df['x₁'] = df['D'] / df['E']
df['x₁'] = df['x₁'].round(2)

df['x₂'] = df['D'] / df['D+E'] * df['Rd'] * (1 - df['T']) + df['E'] / df['D+E'] * df['Re']
df['x₂'] = df['x₂'].round(2)

df['y'] = df['Market'] / df['Nominal']
df['y'] = df['y'].round(2)

df_new = df[['Name', 'x₁', 'x₂', 'y']]
print(df_new)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
# Prepare the features (independent variables) and target (dependent variable)
X = df[['x₁', 'x₂']]  # Independent variables
y = df['y']  # Dependent variable (target)
# Create the regression model
model = LinearRegression()
# Fit the model
model.fit(X, y)
# Get the model coefficients (how each feature affects the dependent variable)
print('Coefficients:', model.coef_)
# Get the intercept (baseline value when x₁ and x₂ are zero)
print('Intercept:', model.intercept_)
# Make predictions based on the fitted model
y_pred = model.predict(X)
# Print the predictions
print('Predictions:', y_pred)
```

```python
import pandas as pd
import statsmodels.api as sm

# Assuming df has the necessary columns: y (dependent), x₁ and x₂ (independent)
# Prepare the features (independent variables) and target (dependent variable)
X = df[['x₁', 'x₂']]  # Independent variables
y = df['y']  # Dependent variable (target)

# Add a constant to the independent variables (for the intercept in the regression model)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Get the summary of the regression
print(model.summary())
```

```python
X = df[['x₁', 'x₂']]  # Independent variables
X = sm.add_constant(X)  # Intercept uchun
y = df['y']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
df['predicted'] = model.predict(X)
df['residuals'] = df['y'] - df['predicted']
import matplotlib.pyplot as plt

plt.scatter(df['predicted'], df['residuals'])
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


# X va y
X = df[['x₁', 'x₂']]
y = df['y']

# Polynomial features yaratish (2-daraja)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Polynomial Regression model qurish
model_poly = LinearRegression()
model_poly.fit(X_poly, y)

# Grid yasash (vizualizatsiya uchun)
x1_range = np.linspace(df['x₁'].min(), df['x₁'].max(), 30)
x2_range = np.linspace(df['x₂'].min(), df['x₂'].max(), 30)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Grid ma'lumotlarini Polynomial feature ga o‘tkazish
grid_points = np.c_[x1_grid.ravel(), x2_grid.ravel()]
grid_poly = poly.transform(grid_points)

# Bashorat qilingan z qiymatlarini hisoblash
z_grid = model_poly.predict(grid_poly).reshape(x1_grid.shape)

# Grafik
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# Real nuqtalarni chizish
ax.scatter(df['x₁'], df['x₂'], y, color='blue', label='Real y')

# Sirtni chizish
ax.plot_surface(x1_grid, x2_grid, z_grid, alpha=0.5, cmap='plasma')

# O‘qlar nomi
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.set_zlabel('y')

plt.title('3D Polynomial Regression (degree=2)')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Ma'lumot
data = {
    'Name': ["Xalq Banki", "Asia Alliance Bank ATB", "IMKON FINANS AJ", "Ipak Yo'li AITB",
             "Ipoteka-bank ATIB", "Turonbank", "Agrobank ATB", "KDB Bank O'zbekiston",
             "Kapitalbank ATB", "InFinBank AJ", "O'zsanoatqurilishbank ATB", "BRB Bank"],
    'x₁': [2.98, 5.47, 4.00, 5.13, 5.29, 7.40, 4.42, 5.75, 0.17, 7.11, 7.55, 7.51],
    'x₂': [0.12, 0.17, 0.19, 0.01, 0.04, 0.00, 0.00, 0.62, 0.00, 0.03, 0.44, 0.03],
    'y' : [1.00, 1.00, 1.10, 9.49, 0.65, 0.55, 1.00, 5.56, 1.00, 0.47, 4.60, 0.46]
}

df = pd.DataFrame(data)
```

```python
# X va y
X = df[['x₁', 'x₂']]
y = df['y']

# Polynomial features (2-daraja)
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X)

# Model
model_poly = LinearRegression()
model_poly.fit(X_poly, y)

# Koeffitsiyent va intercept
coefficients = model_poly.coef_
intercept = model_poly.intercept_

# Polynomial tenglamasi
feature_names = poly.get_feature_names_out(['x₁', 'x₂'])

for name, coef in zip(feature_names, coefficients):
    print(f"{coef:.4f} * {name}")

print(f"Intercept (b₀): {intercept:.4f}")
```

```python
import statsmodels.api as sm

# X va y
X = df[['x₁', 'x₂']]
y = df['y']

# Constant qo‘shamiz (intercept uchun)
X = sm.add_constant(X)

# OLS model
model_ols = sm.OLS(y, X).fit()

# Natijani chiqaramiz
print(model_ols.summary())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt

# Ma'lumotlarni kiritish
data = {
    'x1': [2.98, 5.47, 4.00, 5.13, 5.29, 7.40, 4.42, 5.75, 0.17, 7.11, 7.55, 7.51],
    'x2': [0.12, 0.17, 0.19, 0.01, 0.04, 0.00, 0.00, 0.62, 0.00, 0.03, 0.44, 0.03],
    'y':  [1.00, 1.00, 1.10, 9.49, 0.65, 0.55, 1.00, 5.56, 1.00, 0.47, 4.60, 0.46]
}
df = pd.DataFrame(data)

# Outlierlarni Z-score orqali aniqlash (Z > 2.5 bo'lsa outlier)
z_scores = np.abs(zscore(df))
df_no_outliers = df[(z_scores < 2.5).all(axis=1)]

# Korrelyatsiya matritsiyasi
corr_matrix_no_outliers = df_no_outliers.corr()

# Natijani ko'rsatish
print("Outlierlarsiz korrelyatsiya matritsiyasi:")
print(corr_matrix_no_outliers)

# Heatmap chizish
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix_no_outliers, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korrelyatsiya matritsiyasi (outlierlarsiz)")
plt.show()
```
