```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2020-2024 yil uchun taxminiy ma'lumotlar (trillion USD)
data = {
    "Year": [2020, 2021, 2022, 2023, 2024],
    "Revenue": [3.42, 4.05, 4.90, 4.44, 4.90],
    "Total_Outlays": [6.55, 6.82, 6.27, 6.11, 6.50],
    "Social_Security": [1.30, 1.33, 1.35, 1.38, 1.40],
    "Medicare": [0.68, 0.70, 0.72, 0.73, 0.75],
    "Medicaid": [0.45, 0.46, 0.49, 0.50, 0.52],
    "Defense": [0.72, 0.74, 0.78, 0.86, 0.88],
    "Education": [0.12, 0.17, 0.18, 0.18, 0.18],
    "Welfare_Other": [0.40, 0.40, 0.44, 0.40, 0.42],
    "Interest_on_Debt": [0.35, 0.40, 0.45, 0.48, 0.50],
}

df = pd.DataFrame(data)
df["Deficit"] = df["Total_Outlays"] - df["Revenue"]

# Stil sozlash
sns.set_style("whitegrid")

# 1 Har bir xarajat kategoriyasi trendi
plt.figure(figsize=(12,6))
for col in ["Social_Security","Medicare","Medicaid","Defense","Education","Welfare_Other","Interest_on_Debt"]:
    plt.plot(df["Year"], df[col], marker='o', label=col)

plt.title("2020-2024: US Federal Budget Components (trillion USD)")
plt.xlabel("Year")
plt.ylabel("Spending (trillion USD)")
plt.legend()
plt.show()

# 2 Revenue va Total Outlays + Deficit
plt.figure(figsize=(12,6))
plt.plot(df["Year"], df["Revenue"], marker='o', label="Revenue")
plt.plot(df["Year"], df["Total_Outlays"], marker='o', label="Total Outlays")
plt.bar(df["Year"], df["Deficit"], alpha=0.3, color='red', label="Deficit")
plt.title("2020-2024: Revenue vs Total Outlays and Deficit")
plt.xlabel("Year")
plt.ylabel("Trillion USD")
plt.legend()
plt.show()
```
