```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Your data
data = {
    'Year': [2024, 2024, 2024, 2024, 2024],
    'Company_Name': ['Apple', 'Microsoft', 'Unilever', 'Toyota', 'Google'],
    'Industry': ['Technology', 'Technology', 'Consumer Goods', 'Automotive', 'Technology'],
    'Total_Assets': [364980, 512160, 86300, 624800, 450260],
    'Total_Debt': [96660, 62220, 31900, 252100, 28500],
    'Net_Profit': [93740, 88140, 6220, 25200, 100120],
    'ROA': [0.26, 0.17, 0.07, 0.04, 0.22],
    'ROE': [0.35, 0.20, 0.11, 0.07, 0.24],
    'Sustainable_Procurement': [1, 1, 0, 0, 1],
    'Procurement_Amount': [15000, 12000, 7000, 8000, 14000],
    'Tender_Count': [220, 180, 100, 150, 210],
    'ESG_Score': [85, 88, 75, 78, 90]
}

# Create DataFrame
df = pd.DataFrame(data)

# Log transformation for Total_Assets and Procurement_Amount
df['Log_Total_Assets'] = df['Total_Assets'].apply(lambda x: np.log(x))
df['Log_Procurement_Amount'] = df['Procurement_Amount'].apply(lambda x: np.log(x))

# Define independent and dependent variables
X = df[['Sustainable_Procurement', 'Log_Total_Assets', 'Total_Debt', 'Log_Procurement_Amount', 'Tender_Count', 'ESG_Score']]
Y = df['ROA']

# Add constant to the model (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X).fit()

# Display the results
print(model.summary())
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Step 1: Create your dataset
data = {
    'Year': [2024, 2024, 2024, 2024, 2024],
    'Company_Name': ['Apple', 'Microsoft', 'Unilever', 'Toyota', 'Google'],
    'Industry': ['Technology', 'Technology', 'Consumer Goods', 'Automotive', 'Technology'],
    'Total_Assets': [364980, 512160, 86300, 624800, 450260],
    'Total_Debt': [96660, 62220, 31900, 252100, 28500],
    'Net_Profit': [93740, 88140, 6220, 25200, 100120],
    'ROA': [0.26, 0.17, 0.07, 0.04, 0.22],
    'ROE': [0.35, 0.20, 0.11, 0.07, 0.24],
    'Sustainable_Procurement': [1, 1, 0, 0, 1],
    'Procurement_Amount': [15000, 12000, 7000, 8000, 14000],
    'Tender_Count': [220, 180, 100, 150, 210],
    'ESG_Score': [85, 88, 75, 78, 90]
}

df = pd.DataFrame(data)

# Step 2: Add log-transformed variables
df['Log_Total_Assets'] = np.log(df['Total_Assets'])
df['Log_Procurement_Amount'] = np.log(df['Procurement_Amount'])

# Step 3: Define independent variables (X)
X = df[['Sustainable_Procurement', 'Log_Total_Assets', 'Total_Debt', 
        'Log_Procurement_Amount', 'Tender_Count', 'ESG_Score']]
X = sm.add_constant(X)  # add intercept

# Step 4: Run OLS Regression for ROA
model_roa = sm.OLS(df['ROA'], X).fit()
print("======= OLS Regression for ROA =======")
print(model_roa.summary())

# Step 5: Run OLS Regression for ROE
model_roe = sm.OLS(df['ROE'], X).fit()
print("\n======= OLS Regression for ROE =======")
print(model_roe.summary())
```
