## Data vizualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('movie.csv')
```

```python
df.head()
```

```python
#Line plot
plt.plot(df['num_critic_for_reviews'], df['duration'])
plt.title('Movie')
plt.xlabel('Reviews')
plt.ylabel('duration')
plt.show()
```

```python
#Line plot
plt.scatter(df['num_critic_for_reviews'], df['duration'])
plt.title('Movie')
plt.xlabel('Reviews')
plt.ylabel('duration')
plt.show()
```

```python
# Fix for the bar plot
df_cleaned = df.dropna(subset=['country', 'duration'])  # Remove rows with NaN in 'country' or 'duration'
country_means = df_cleaned.groupby('country')['duration'].mean()

plt.bar(country_means.index, country_means.values)
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.title('Average Movie Duration by Country')
plt.xlabel('Country')
plt.ylabel('Average Duration')
plt.show()
```
